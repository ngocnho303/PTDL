```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings("ignore")

plt.rcParams['font.family'] = 'sans-serif'

# đọc file dữ liệu
df = pd.read_csv("datasetV2.csv")
# đổi tên cột sang tiếng việt
new_names = {
    'Order_ID': 'ID_DonHang',
    'Customer_ID': 'ID_KhachHang',
    'Date': 'Ngay_Mua',
    'Age': 'Tuoi',
    'Gender': 'Gioi_Tinh',
    'City': 'Thanh_Pho', 
    'Product_Category': 'Loai_San_Pham',
    'Unit_Price': 'Don_Gia',
    'Quantity': 'So_Luong',
    'Discount_Amount': 'Tien_Giam_Gia',
    'Total_Amount': 'Tong_Chi_Tieu',          
    'Session_Duration_Minutes': 'TG_Truy_Cap',
    'Pages_Viewed': 'So_Trang_Xem',
    'Customer_Rating': 'Danh_Gia',
    'Delivery_Time_Days': 'Thoi_Gian_Giao_Hang',
    'Payment_Method': 'Phuong_Thuc_Thanh_Toan',
    'Device_Type': 'Loai_Thiet_Bi',
    'Is_Returning_Customer': 'Khach_Hang_Quay_Lai'
}
df.rename(columns=new_names, inplace=True)
df.head(10)

df['Ngay_Mua'] = pd.to_datetime(df['Ngay_Mua'])
#Hiển thị các cột ban đầu 
print('Các cột trong dataset:\n', df.columns.tolist())
#Ktra ttin
df.info() #Lấy thông tin về số lượng và kiểu dữ liệu của các cột trong DataFrame
print( 'Số dòng và cột của bộ dữ liệu:',df.shape)
df.head(10) # Hiển thị trước 10 hàng đầu tiên của tập dữ liệu
# Kết luận  =>Đây là bộ dl khá tốt không có mising value ở các biến 
print(" Tổng số dòng dữ liệu:", len(df))

Ngay_Hien_Tai = df['Ngay_Mua'].max() + pd.Timedelta(days=1)


#Tính toán các thông số thống kê của bộ dữ liệu ( chỉ có thuộc tính number mới được tính toán )
df.describe()

#Thống kê cho các cột dữ liệu chuỗi (object)
df.describe(include=['object'])

# Đếm số lượng giá trị duy nhất của mỗi cột
print("Đếm số lượng giá trị duy nhất trong mỗi cột:")
print(df.nunique())


#Kiểm tra dữ liệu bị thiếu
print(df.isnull().sum())  # Đếm số lượng missing value từng cột
print('Số lượng dòng trùng :',df.duplicated().sum())  # Đếm số lượng dòng trùng

cutoff_date = df['Ngay_Mua'].quantile(0.7)

train_df = df[df['Ngay_Mua'] <= cutoff_date].copy()
test_df  = df[df['Ngay_Mua'] > cutoff_date].copy()

print(f" Train: {train_df['Ngay_Mua'].min().date()} → {train_df['Ngay_Mua'].max().date()}")
print(f" Test : {test_df['Ngay_Mua'].min().date()} → {test_df['Ngay_Mua'].max().date()}")



def build_customer_features(data, snapshot):
    rfm = data.groupby('ID_KhachHang').agg({
        'Ngay_Mua': lambda x: (snapshot - x.max()).days,
        'ID_DonHang': 'count',
        'Tong_Chi_Tieu': 'sum',
        'TG_Truy_Cap': 'mean',
        'So_Trang_Xem': 'mean'
    }).reset_index()

    rfm.columns = [
        'ID_KhachHang',
        'Do_Moi',
        'Tan_Suat',
        'Tong_Chi_Tieu',
        'TG_Truy_Cap_TB',
        'So_Trang_Xem_TB'
    ]
    return rfm

train_rfm = build_customer_features(train_df,Ngay_Hien_Tai )
test_rfm  = build_customer_features(test_df, Ngay_Hien_Tai)

features_vn = [
    'Do_Moi',
    'Tan_Suat',
    'Tong_Chi_Tieu',
    'TG_Truy_Cap_TB',
    'So_Trang_Xem_TB'
]

train_rfm[features_vn] = train_rfm[features_vn].fillna(train_rfm[features_vn].median())
test_rfm[features_vn]  = test_rfm[features_vn].fillna(test_rfm[features_vn].median())

# Scale dữ liệu
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train_rfm[features_vn])

# Tìm số cụm bằng phương pháp elbow và silhotte
inertias, silhouettes = [], []
K_range = range(2, 8)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_tmp = km.fit_predict(X_train_scaled)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_train_scaled, labels_tmp))

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(K_range, inertias, marker='o')
plt.axvline(3, color='red', linestyle='--')
plt.title("Elbow Method")

plt.subplot(1,2,2)
plt.plot(K_range, silhouettes, marker='o')
plt.axvline(3, color='red', linestyle='--')
plt.title("Silhouette Score")
plt.show()

# Chọn số cụm = 3 dựa vào 2 phương pháp trên
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
train_rfm['Cum'] = kmeans.fit_predict(X_train_scaled)

X_test_scaled = scaler.transform(test_rfm[features_vn])
test_rfm['Cum'] = kmeans.predict(X_test_scaled)
# Gán tên phân khúc khách hàng 
cluster_order = (
    train_rfm
    .groupby('Cum')['Tong_Chi_Tieu']
    .mean()
    .sort_values()
    .index
)

segment_map = {
    cluster_order[0]: 'Giá trị thấp (Vãng lai)',
    cluster_order[1]: 'Giá trị trung bình (Tiềm năng)',
    cluster_order[2]: 'Giá trị cao (VIP)'
}

train_rfm['Phan_Khuc'] = train_rfm['Cum'].map(segment_map)
test_rfm['Phan_Khuc']  = test_rfm['Cum'].map(segment_map)

print("\n         Phân bố phân khúc (Train):")
print(train_rfm['Phan_Khuc'].value_counts())



# Thống kê số lượng khách hàng theo từng cụm
so_luong_theo_cum_df = (
    train_rfm
    .groupby(['Cum', 'Phan_Khuc'])
    .size()
    .reset_index(name='So_Luong_Khach_Hang')
)

print("Số lượng khách hàng theo từng cụm:")
so_luong_theo_cum_df.head(5)


# Thống kê số lượng khách hàng theo từng cụm
so_luong_theo_cum_df = (
    train_rfm
    .groupby(['Cum', 'Phan_Khuc'])
    .size()
    .reset_index()
)

print("Số lượng khách hàng theo từng cụm:")
so_luong_theo_cum_df.head(10)

# Trung bình các biến theo từng cụm
TB_Bien_Cum_df = (
    train_rfm
    .groupby(['Cum', 'Phan_Khuc'])[
        ['Do_Moi', 'Tan_Suat', 'Tong_Chi_Tieu', 'TG_Truy_Cap_TB', 'So_Trang_Xem_TB']
    ]
    .mean()
    .round(2)
    .reset_index()
)

TB_Bien_Cum_df.head(5)


plt.figure(figsize=(6,4))
sns.countplot(
    x='Phan_Khuc',
    data=train_rfm,
    order=train_rfm['Phan_Khuc'].value_counts().index
)
plt.title("Phân bố khách hàng theo phân khúc")
plt.xticks(rotation=15)
plt.show()

# Biểu đồ cột so sánh đặc ddiemr trung bình
summary_long = (
    train_rfm
    .groupby('Phan_Khuc')[features_vn]
    .mean()
    .reset_index()
    .melt(id_vars='Phan_Khuc', var_name='Chi_Tieu', value_name='Gia_Tri_TB')
)

plt.figure(figsize=(10,5))
sns.barplot(
    data=summary_long,
    x='Chi_Tieu',
    y='Gia_Tri_TB',
    hue='Phan_Khuc'
)

plt.title("So sánh đặc điểm trung bình của khách hàng theo phân khúc")
plt.xlabel("Chỉ tiêu")
plt.ylabel("Giá trị trung bình")
plt.xticks(rotation=15)
plt.legend(title="Phân khúc")
plt.tight_layout()
plt.show()



from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train_scaled)

pca_df = pd.DataFrame({
    'PC1': X_pca[:, 0],
    'PC2': X_pca[:, 1],
    'Phan_Khuc': train_rfm['Phan_Khuc']
})

plt.figure(figsize=(9,7))
sns.scatterplot(
    data=pca_df,
    x='PC1',
    y='PC2',
    hue='Phan_Khuc',
    alpha=0.7
)

plt.title("Trực quan hóa phân khúc khách hàng (PCA)")
plt.xlabel("Thành phần chính 1")
plt.ylabel("Thành phần chính 2")
plt.legend(title="Phân khúc")
plt.tight_layout()
plt.show()

X_train = train_rfm[features_vn]
y_train = train_rfm['Cum']
X_test  = test_rfm[features_vn]
y_test  = test_rfm['Cum']

rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42,
    class_weight='balanced'
)

rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

from sklearn.metrics import accuracy_score

do_chinh_xac = accuracy_score(y_test, y_pred)
print(f"   Độ chính xác của mô hình (Accuracy): {do_chinh_xac:.4f}")

print("\n  BÁO CÁO ĐÁNH GIÁ MÔ HÌNH\n")
unique_labels = np.unique(y_test)
display_labels = [segment_map[i] for i in unique_labels]
print(classification_report(
    y_test,
    y_pred,
    labels=unique_labels,
    target_names=display_labels
))


y_pred_train = rf_model.predict(X_train)

acc_train = accuracy_score(y_train, y_pred_train)
acc_test  = accuracy_score(y_test, y_pred)

print("\n ĐÁNH GIÁ ĐỘ CHÍNH XÁC MÔ HÌNH")
print(f"  Accuracy trên tập Train: {acc_train:.4f}")
print(f"  Accuracy trên tập Test : {acc_test:.4f}")



# Confusion matrix ( ma trận nhầm lẫn )
unique_labels = np.unique(y_test)

cm = confusion_matrix(y_test, y_pred, labels=unique_labels)
display_labels = [segment_map[i] for i in unique_labels]

plt.figure(figsize=(8,6))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=display_labels,
    yticklabels=display_labels
)
plt.xlabel("Dự đoán")
plt.ylabel("Thực tế")
plt.title("Ma trận nhầm lẫn phân khúc khách hàng")
plt.show()


importance = (
    pd.Series(rf_model.feature_importances_, index=features_vn)
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))
sns.barplot(x=importance.values, y=importance.index)
plt.title("Mức độ quan trọng của các đặc trưng")
plt.show()

def du_doan_khach_hang(do_moi, tan_suat, tong_chi_tieu,
                       tg_tb=15, so_trang_tb=8):
    data = pd.DataFrame(
        [[do_moi, tan_suat, tong_chi_tieu, tg_tb, so_trang_tb]],
        columns=features_vn
    )
    cum = rf_model.predict(data)[0]
    print(" Phân khúc khách hàng:", segment_map[cum])
# Test thử
du_doan_khach_hang(5, 20, 20000)
du_doan_khach_hang(150, 1, 200)


train_rfm.to_csv("ket_qua_phan_khuc_train.csv", index=False, encoding='utf-8-sig')
test_rfm.to_csv("ket_qua_phan_khuc_test.csv", index=False, encoding='utf-8-sig')

print("\n HOÀN THÀNH TOÀN BỘ PHÂN TÍCH")

```
